```python
import llm2geneset
import json
import time
import openai
```

```python
# Open the file in read mode
with open('hgnc_protein_symbols.txt', 'r') as file:
    # Read all lines into a list, stripping the newline character
    hgnc = [line.strip() for line in file]

# Print the list to verify the content
len(hgnc)
```

```python
models = ["gpt-3.5-turbo-0125", "gpt-4o-2024-05-13"]
models = ["gpt-3.5-turbo-0125"]
lib_names = ["KEGG_2021_Human", 
             "Reactome_2022", 
             "WikiPathway_2023_Human"]
lib_names = ["KEGG_2021_Human"]
```

```python
# Generate gene sets for ensembling.
aclient = openai.AsyncClient()
for model in models:
    for lib_name in lib_names:
        with open("libs_human/" + model + "/" + lib_name + ".json") as f:
            gen_res = json.load(f)
                    
        res = await llm2geneset.get_genes_list(aclient, hgnc,
                                               gen_res["descr_cleaned"][0],
                                               model=model)
        
        #with open('libs_human/' + model + '/' + lib_name + '.json', 'w') as json_file:
            #json.dump(gen_res, json_file, indent=4)
```
